```python
from __future__ import division
import os, sys,csv, glob
import numpy as np
import pandas as pd
from collections import defaultdict
from python_latexify import latexify, format_axes
latexify()

import matplotlib
%matplotlib nbagg
matplotlib.rcParams.update(latexify())
import matplotlib.pyplot as plt
```

```python
def getCDF(data):
    xdata = np.sort(data)
    ydata = [i/len(xdata) for i in range(len(xdata))]
    return xdata, ydata

```

```python
## number of countries that hosts the domain:
```

```python
domain_pop = pd.read_csv('hist_data_US.txt',header=None, delimiter="\n",
                     names = ['freq'])
```

```python
x,y =getCDF(list(domain_pop['freq']))
```

```python
fig1, ax1 = plt.subplots(1,1,figsize=(5,3))

ax1.plot(x,y,label='Diff b/w 75th percentile',color='black', lw=2) 

ax1.set_xlabel("Number of Countries that Host a Domain",fontsize=14)
ax1.set_ylabel("Fraction of Domains",fontsize=14)

ax1.grid(1)
for tick in ax1.get_yticklabels():
    tick.set_fontsize(14)
for tick in ax1.get_xticklabels():
    tick.set_fontsize(14) 
fig1.tight_layout()

ax1.annotate('Single Location',xy=(1.1, 0.39),
            xytext=(0.25, 0.2), textcoords='axes fraction',
            arrowprops=dict(facecolor='orange', shrink=0.005),
            fontsize=14
            )

ax1.annotate('CDNs',xy=(17, 0.85),
            xytext=(0.8, 0.2), textcoords='axes fraction',
            arrowprops=dict(facecolor='orange', shrink=0.005),
            fontsize=14
            )

fig1.savefig('domain_hist_US1.pdf')
```

```python
data_nh = pd.read_csv('netherlands_trombone_results.txt',header=None,delimiter=',', names = ['code','count','frac'])
data_nh.sort_values(by=['frac'],ascending=1,inplace=True)
data_kn = pd.read_csv('kenya_trombone_results.txt',header=None,delimiter=',', names = ['code','count','frac'])
data_kn.sort_values(by=['frac'],ascending=1,inplace=True)
data_br = pd.read_csv('brazil_trombone_results.txt',header=None,delimiter=',', names = ['code','count','frac'])
data_br.sort_values(by=['frac'],ascending=1,inplace=True)
```

```python
fig1, ax1 = plt.subplots(1,1,figsize=(5,3))
ax1.bar(range(len(data_nh)),data_nh['frac'],color='orange') 
plt.xticks(map(lambda x: x + 0.5,range(len(data_nh))),data_nh['code'])
ax1.set_ylabel("Fraction of Paths",fontsize=14)

ax1.grid(1)
for tick in ax1.get_yticklabels():
    tick.set_fontsize(14)
for tick in ax1.get_xticklabels():
    tick.set_fontsize(12) 
fig1.tight_layout()
ax1.set_xlim([0, 11])
ax1.set_ylim([0, 1])
fig1.savefig('nl_trombone_new1.pdf')
```
